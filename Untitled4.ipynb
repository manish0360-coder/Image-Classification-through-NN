```python
import tensorflow as tf
import numpy as np
import PIL.Image
import matplotlib.pyplot as plt

# Load pre-trained model
base_model = tf.keras.applications.InceptionV3(include_top=False, weights='imagenet')

# Choose layers to dream on
layer_names = ['mixed3', 'mixed5', 'mixed7']
layers = [base_model.get_layer(name).output for name in layer_names]

dream_model = tf.keras.Model(inputs=base_model.input, outputs=layers)

# Load image
from PIL import Image, ImageOps

def load_image(path):
    img = Image.open(path)
    img = ImageOps.exif_transpose(img)   # <<< THIS LINE FIXES ROTATION
    img = np.array(img)
    img = tf.image.resize(img, (224, 224))
    img = tf.cast(img, tf.float32)
    img = tf.expand_dims(img, axis=0)
    return img

# DeepDream function
def deep_dream(img, steps=300, step_size=0.05):
    img = tf.Variable(img)
    for i in range(steps):
        with tf.GradientTape() as tape:
            outputs = dream_model(img)
            loss = tf.reduce_sum(outputs)
        grads = tape.gradient(loss, img)
        grads = grads / (tf.math.reduce_std(grads) + 1e-8)
        img.assign_add(grads * step_size)
    return img

# Upload your image
from google.colab import files
uploaded = files.upload()
path = list(uploaded.keys())[0]

img = load_image(path)
dreamed = deep_dream(img)

# Show result
result = np.clip(dreamed[0] / 255, 0, 1)
plt.imshow(result)
plt.axis('off')
```
